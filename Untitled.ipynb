```python
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
wine_dataset = datasets.load_wine
help(wine_dataset)
```

```python
wine_dict = wine_dataset()
wine_targets = pd.DataFrame(data=wine_dict['target'],
                            columns=['region'])
# for the sake of fun, i will put the region names instead of the int64 values.
# just so that i ca practice encoding text/categorical attributes.
wine_targets = wine_targets.replace([0, 1, 2], 
                                    [wine_dict['target_names'][0], wine_dict['target_names'][1],wine_dict['target_names'][2]])

wine_data = pd.DataFrame(data=wine_dict['data'], 
                         columns=wine_dict['feature_names'])

wine_data = pd.concat([wine_data, wine_targets], axis=1)
```

```python
# I will skip StratifiedShuffleSplit since i dont know what could make a good
# stratification strategy. I don't think there is a need to have the data stratified.
# This assumption could of course be incorrect, but i am willing to take the risk.

from zlib import crc32

def test_set_check(id_, test_ratio):
    return crc32(np.int64(id_)) & 0xffffffff < test_ratio * 2**32


def split_train_test(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio))
    return data.loc[~in_test_set], data[in_test_set]

wine_data = wine_data.reset_index()
wine_data
train_set, test_set = split_train_test(wine_data, 0.2, 'index')
wine_data = wine_data.drop(['index'], axis=1)
for set_ in (train_set, test_set):
    set_.drop('index', axis=1, inplace=True)
```
