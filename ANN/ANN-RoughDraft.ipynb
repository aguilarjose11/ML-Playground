```python
from ANN import Network
from ANNNetwork import Network as ANNE

from sklearn import datasets
import pandas as pd
import numpy as np

from typing import List
```

```python
digits = datasets.load_digits()
```

```python
print(digits.DESCR)
```

```python
def MINST_label_encoder(data: np.ndarray) -> np.ndarray:
    # Encode in a one-hot manner
    def _digit_encoder(datum: int)->List[int]:
        val = np.zeros(10)
        val[datum] = 1
        return val
    return [_digit_encoder(datum).reshape((10, 1)) for datum in data]
            
```

```python
MINST = [ datum.reshape((64, 1)) for datum in digits.data]
```

```python
MINST[0].shape
```

```python
MINST_labels = MINST_label_encoder(digits.target)
```

# Kaggle's MINST

```python
MINST_kaggle = pd.read_csv("./train.csv")
MINST_kaggle_y = MINST_label_encoder(MINST_kaggle["label"].to_numpy())
MINST_kaggle_x = MINST_kaggle.drop("label", axis=1).to_numpy()
MINST_kaggle_x = [ datum.reshape((784, 1)) for datum in MINST_kaggle_x]
```
