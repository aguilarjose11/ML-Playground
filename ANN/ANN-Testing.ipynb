```python
from ANN import Network
from ANNNetwork import Network as ANNE

from sklearn.preprocessing import Normalizer
from sklearn import datasets
import pandas as pd
import numpy as np

from typing import List
```

```python
def MINST_label_encoder(data: np.ndarray) -> np.ndarray:
    # Encode in a one-hot manner
    def _digit_encoder(datum: int)->List[int]:
        val = np.zeros(10)
        val[datum] = 1
        return val
    return [_digit_encoder(datum).reshape((10, 1)) for datum in data]
```

```python
MINST_kaggle_raw = pd.read_csv("./train.csv").sample(frac=1).reset_index(drop=True)

normalizer = Normalizer()

MINST_kaggle_y = MINST_label_encoder(MINST_kaggle_raw["label"].to_numpy())
MINST_kaggle_x_raw = MINST_kaggle_raw.drop("label", axis=1)#.to_numpy()
MINST_kaggle_x_normal = normalizer.fit_transform(MINST_kaggle_x_raw)
MINST_kaggle_x = [ datum.reshape((784, 1)) for datum in MINST_kaggle_x_normal]

MINST_kaggle = []
for i in range(len(MINST_kaggle_y)):
    MINST_kaggle.append([MINST_kaggle_x[i], MINST_kaggle_y[i]])
```
